```python
# импорт библиотеки
from scipy.stats import norm

# norm - объект, реализующий нормальное распределение
# полный список - https://docs.scipy.org/doc/scipy/reference/stats.html.
```

```python
# pdf - метод для вычисления плотности вероятностей заданного распределения
# cdf - метод для вычисления функции распределения

```

```python
# значение плотного стандартного нормального распределения N(0,1) в точке х
x = 1
norm.pdf(x)
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

xx = np.linspace(-2, 4, 101)
loc_ = 1
scale_ = 1
plt.plot(xx, norm.pdf(xx, loc=loc_, scale=scale_), label=f"loc={loc_}, scale={scale_}")

loc_ = 0
scale_ = 1
plt.plot(xx, norm.pdf(xx, loc=loc_, scale=scale_), label=f"loc={loc_}, scale={scale_}")

loc_ = 1
scale_ = 0.5
plt.plot(xx, norm.pdf(xx, loc=loc_, scale=scale_), label=f"loc={loc_}, scale={scale_}")

loc_ = 1
scale_ = 2
plt.plot(xx, norm.pdf(xx, loc=loc_, scale=scale_), label=f"loc={loc_}, scale={scale_}")

plt.grid()
plt.legend(loc="best")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min
from numpy import exp

x = np.linspace(0, 3, 101)
r = 2
scale = 2

plt.plot(x, weibull_min.pdf(x, c=r, scale=scale));
```

```python
x = np.linspace(0, 3, 101)
r = 2
scale = 2

plt.plot(x, weibull_min.pdf(x, c=r, scale=scale));
plt.plot(x, r * x**(r-1) * exp(-(x/scale)**r) / (scale**r), lw=4, alpha=0.4)
```
